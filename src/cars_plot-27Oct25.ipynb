```python
import polars as pl
import altair as alt
```

```python
df = pl.read_csv('data/cars.csv', ignore_errors=True)
```

```python
df = df.drop_nulls(subset='mpg')
```

```python
df['model year'].min()
```

```python
alt.Chart(df).mark_bar().encode(alt.X('horsepower:Q', bin=True), y='count()')
```

```python
df.group_by('model year').agg(pl.len()).plot.bar(x='model year', y='len')
```

```python
df.group_by('model year').mean().plot.line(x='model year', y='mpg')
```
