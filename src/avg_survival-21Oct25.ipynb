```python
import pandas as pd

df = pd.read_csv('ddi13-warmups/src/data/titanic_survival.csv')
```

```python
df.head()
```

```python
first, second, third = df.groupby('class').mean(numeric_only=True)['survived']
```

```python
fem_num, male_num = df['sex'].value_counts().iloc[:]
```

```python
df['child'] = df['age'] < 18
```

```python
child_num = df['child'].count()
```

```python
totals = {'first_class': first, 'second_class': second, 'third_class:' third, 'female': fem_num, 'male': male_num, 'child': child_num}
print(totals)
```
